# Importing libraries

```python
#Import libraries.
import sklearn
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading and Cleaning data.

```python
df = pd.read_csv(r"India_Tourism_2014_2020.csv") #Load file
```

```python
#Get information of data and understand data.
df.describe()
df.info()
df
```

```python
#Find nulls.
df.isnull().sum()
df = df.fillna(0) #Repllace all nulls in rows with 0. (0 visited from country)

#Removing duplicates.
duplicate_count = df.drop_duplicates()
print(duplicate_count)
print()
print(df.duplicated().sum())
```

```python
#Find string type columns. #Only 1. 
cat_cols = df.select_dtypes(include=['object'])
print(cat_cols)
#Capitalizing data.
df['Country of Nationality'] = (df['Country of Nationality'].str.strip().str.capitalize())
```

```python
#Removing negative visits or tours.
num_cols = df.select_dtypes(include="number").columns
df = df[(df[num_cols] >= 0).all(axis=1)]

#Capitalzie every word in country.
df['Country of Nationality'] = df['Country of Nationality'].str.title()
df.columns
# df = df.drop(columns=["Unnamed: 26","Unnamed: 28","Unnamed: 30"]) #Removing some unnamed columns.
```

# Encoding

```python
le = LabelEncoder()
cat_cols = df.select_dtypes(include=["object"])
print(cat_cols)
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

# Model Building

Linear Regression Model

```python
df2 = df
from sklearn.linear_model import LinearRegression
```

```python
#Define independent(x) and dependent(y) vairaibles.
x = df2.drop(["Country of Nationality", "2020 4th quarter (Oct-Dec) "], axis = 1)
y = df2["2020 4th quarter (Oct-Dec) "]
```

```python
#Train model through variables.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
#Traing model on 80% of data.
model = LinearRegression() #Creating instance of function.
model.fit(x_train,y_train)
```

```python
#Testing model errors on 20% left data.
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
MAE = mean_absolute_error(y_test,y_pred) 
MSE = mean_squared_error(y_test,y_pred)
R2_Score = r2_score(y_test,y_pred)
print(MAE)
print(MSE)
print("Accuracy:",R2_Score)
```

### Using for loop run all models.

```python
# from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {"Logistic Regression":LogisticRegression(), "Decision Tree":DecisionTreeClassifier(random_state=42),"Random Forest": RandomForestClassifier(random_state=42), "KNN":KNeighborsClassifier(n_neighbors=3)}
#dict to pick and iterate through models.
```

```python
best_model = None
best_accuracy = 0
from sklearn.metrics import accuracy_score
```

### If we want to add Linear Regression we can import all the commented libs.

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    
    #Add this below for linear regression.
    #  if name == "Linear Regression":
    #     y_pred = np.round(y_pred).astype(int)
    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Name:{name} and Accuracy:{accuracy}")
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model
        best_model_name = name
```

```python
print("Model with best accuracy:",best_model_name)
print("Accuracy:",best_accuracy)
```
