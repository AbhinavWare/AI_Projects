### Step 4 : Data cleaning and pre-proccessing. Dealing with inconsistant values.

Step 4.1: Data case normalization

```python
#Check unique values in columns.
print(df['Gender'].unique()) #[] pass column name to perform operation on it.
#Clean this data.
df['Gender']=(df['Gender'].str.strip().str.capitalize()) #str.strip() removes leading and trailing whitespace, str.capitalize() capitalizes the first letter of each word.
print(df['Gender'].unique()) 
#Same for Internship column.
df['Internship']=(df['Internship'].str.strip().str.capitalize()) #str.strip() removes leading and trailing whitespace, str.capitalize() capitalizes the first letter of each word.
print(df['Internship'].unique()) 
```

Step 4.2 : Finding missing null values.

```python
print(df.isnull().sum()) #Prints the number of missing values in each column of the data frame.
```

Step 4.3 : Filling null cells.

```python
df['CGPA'] = df['CGPA'].fillna(df['CGPA'].mean())
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].mean())
df['Certifications'] = df['Certifications'].fillna(df['Certifications'].median())
df['Internship'] = df['Internship'].fillna(df['Internship'].mode()[0])
df['Salary_LPA'] = df['Salary_LPA'].fillna(0)

print(df.isnull().sum())
df
```

Step 4.4 : Removing Duplicates.

```python
duplicate_count = df.drop_duplicates().sum()
print(duplicate_count)
print()
print(df.duplicated().sum())
```

Step 4.5 : Find and solve out-of-range values.

```python
print(df[df['CGPA'] > 10]) #Check CGPA above 10.
#To change.
df.loc[df['CGPA'] > 10, "CGPA"] = 10 #Find location of cells having valse greater than 10 using loc() and exchange them with 10.

print(df[df['Attendance'] > 100]) #Check Attendence above 100.
df.loc[df['Attendance'] > 100,"Attendance"] = 100 #We write "Column name" to specify change in only "" column.

df.loc[df['Aptitude_Score'] > 100,"Aptitude_Score"] = 100
df.describe()
```

### Step 6 : Encoding our data to binary so machine can understand.

Label encode. 
Import sklearn libray from scikit-learn.

Creating instance of function.

```python
le = LabelEncoder()
cat_cols = df.select_dtypes(include=['object'])
print(cat_cols)
cat_col = ["Gender","Internship","Department","City","Placement"]
for col in cat_col:
    df[col] = le.fit_transform(df[col])
df
```

### Step 7 : Model Building.

#### Step 1 : Linear Regression model.
#### Model Building Steps : (Commen)
Step 1 : Define dependent and independent variables to find relation and predict output.

Define Independent and dependent variable x and y.

Step 2 : Drop higly co-related columns. So data is not biased.

```python
df_d = df.drop("Placement",axis = 1) #Drop higly co-related column to salary column.
```

```python
x = df_d.drop("Salary_LPA",axis = 1)
y = df_d['Salary_LPA']
```

Step 3 : Divide data for use: 
80% x, y for training.
20% x,y for testing.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42) #Splitting training and testing data.
#random_state is used to increase data accuracy and data split would be different every time.
```

Custom Step 4: Linear Regression model training.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() #Creating instance of function.
model.fit(x_train,y_train)
```

Step 5 : Predictions...

y_test = Actual Value    y_pred = Predicted Value   Error = y_test - y_pred

```python
y_pred = model.predict(x_test)
```

Step 6 : Model Evaluation:

Linear Regression or Regression evaluation ----> Done using MSE,MAE,R Score,RMSE

```python
#We will use R2 Score.
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
#Some Metrics.
MAE = mean_absolute_error(y_test,y_pred) #Average of all errors.
MSE = mean_squared_error(y_test,y_pred) #Big mistakes are made more larger.
R2_Score = r2_score(y_test,y_pred)
print(MAE)
print(MSE)
print(R2_Score)
```
