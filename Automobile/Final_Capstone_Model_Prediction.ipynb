# Cleaning and Pre-processing data.

```python
#Cleaning ? values.
df['normalized-losses'] = df['normalized-losses'].replace("?","df['normalized-losses'].mean()")
df['num-of-doors'] = df['num-of-doors'].replace("?","Two")
df
```

```python
#Checking and Cleaning null values.
print(df.isnull().sum())
#Confirms if null or not.
```

```python
#Capitalize and keeping unique values.
print(df["make"].unique())
df['make']=(df['make'].str.strip().str.capitalize())
print(df['aspiration'].unique())
df['aspiration']=(df['aspiration'].str.strip().str.capitalize()) 
print(df['body-style'].unique())
df['body-style']=(df['body-style'].str.strip().str.capitalize()) 
print(df['drive-wheels'].unique())
df['drive-wheels']=(df['drive-wheels'].str.strip().str.capitalize()) 
print(df['engine-location'].unique())
df['engine-location']=(df['engine-location'].str.strip().str.capitalize()) 
print(df['fuel-type'].unique())
df['fuel-type']=(df['fuel-type'].str.strip().str.capitalize()) 
print(df['num-of-doors'].unique())
df['num-of-doors']=(df['num-of-doors'].str.strip().str.capitalize()) 
```

```python
#Replacing nulls in 'num-of-doors'
df['num-of-doors'].isnull().sum() #found nulls.
df['num-of-doors'] = df['num-of-doors'].fillna(df['num-of-doors'].mode()[0])
df
```

```python
df = df.drop_duplicates()
```

# Model Evaluation.

### Linear Regression model.

```python
#importing libs.
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
#Defining target variables.
x = df.drop("normalized-losses", axis=1) #independent.
y = df["normalized-losses"] #dependent.
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42) #For training & testing.
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
print("Mean absolute error:",mean_absolute_error(y_test,y_pred))
print("Mean squared error:",mean_squared_error(y_test,y_pred))
print("R2 score:",r2_score(y_test,y_pred))
```

### Logistics Regression model.

```python
#importing files.
from sklearn.linear_model import LogisticRegression
```

```python
#We will predict fuel type.
#Remove fuel-type column as it is highly co-related.

#Define variables.
x = df.drop('fuel-type',axis = 1) #independent

y = df['fuel-type']
```

```python
#Train model through variables.
from sklearn.metrics import accuracy_score
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#Variable to store value to predict.
y_pred = model.predict(x_test)
```

```python
acc = accuracy_score(y_test,y_pred)
print("Accuracy score",acc)
```

### Comparing KNN, Decesion tree and Random forest accuracy usinng for loop to  choose most accurate model.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = { "KNN": KNeighborsClassifier(n_neighbors = 3),"Decision Tree": DecisionTreeClassifier(random_state = 42), 
          "Random Forest": RandomForestClassifier(random_state = 42)}
trained_models = {}
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    trained_models[name] = model
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\n{name}")
    print(f"Accuracy: {accuracy:.4f}")
```

### From accuracy score observation we get to know K-nearest neighbour alogorithm provies highest accuracy of 83% 

### For new data points we will use knn to predict accuracy for new car.

```python
new_car = [[1,2,1,3,1,2,1,3,2,1]]
```

```python
pred_knn = trained_models['KNN'].predict(new_car)
pred_knn
```
